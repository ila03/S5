```python
# read web table into pandas DF
import pandas as pd

forestDFs=pd.read_html(ciaLink2, # link
                        header=0, # where is the header?
                        flavor='bs4')
```

```python
# make a copy
forest=forestDFs[0].copy()
```

```python
# here it is
forest
```

```python
#1. Replace '%' by 'pct'
forest1=forest.rename(columns={'% of GDP':'pct of GDP'})
forest1
```

```python
#2. Keep the columns Country, pct of GDP, and Date of Information
forest2=forest1.drop(columns=['Rank'])
forest2
```

```python
#3. Change the column name Date of Information to forest_date.
forest3=forest2.rename(columns={'Date of Information':'forest_date'})
forest3
```

```python
#4. Make sure there are no spaces as part of the column names
forest3['Country']=forest3.Country.str.replace(pat='\s',repl='',regex=True)
forest3
```

```python
#5. Make sure the cells with text does not have neither trailing nor leading spaces
forest3['Country']=forest3.Country.str.strip()
forest3
```

```python
#6. Keep only the year value in the column forest_date.
forest3['forest_date']=forest3.forest_date.str.extract(pat=r'(^\w+|.$)', expand=True)
forest3
```

```python
forestCleaned=forest3
forestCleaned
```
